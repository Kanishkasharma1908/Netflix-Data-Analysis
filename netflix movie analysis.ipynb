#EXPLORATION SUMMARY
* We have a dataset of 8807 rows with 12 columns
* The dataset contains no duplicate values but missing values in columns director, cast, country, date_added, rating, duration
* Date added needs to be casted into date time and to extract only the year value
* The description column is not needed for ananlysis so it will be dropped
* Cast,date_added column has comma seperated values and white spaces which needs to be removed

# Data Cleaning

```python
df['date_added'] = df['date_added'].str.strip()           # To strip the extra white spaces so that this doesnt give error
df['date_added'] = pd.to_datetime(df['date_added'])   
print(df['date_added'].dtype)                       # print(type(['date_added'])) ---> gives you the dtype of the whole column whiich is a pandas series 
```

```python
df['date_added'] = df['date_added'].dt.year.astype('Int64')  # to extract only the year part of date 
print(df['date_added'].dtype)
```

```python
# removing unwanted columns 
df.drop(['director','description'], axis = 1, inplace = True)    # inplace = to permanently remove cols, axis = 1 --> cols, 2 --> rows
df.head()
```

```python
# Labelling the release year column into bins 

df['release_year'] = pd.to_numeric(df['release_year'])   
bins = [1900, 2000, 2010, 2020, 2030]
labels = ['Before 2000s', '2000s', '2010s', 'Recent']
df['release_year'] = pd.cut(df['release_year'], bins = bins, labels = labels)
df.head()
```

```python
df['release_year'].unique()
```

```python
df['release_year'].value_counts()
```

```python
#to remove NA values from all columns 
df.dropna(inplace =  True)
df.isna().sum()
```

```python
df.head()
```

```python
# EXploding the listed_in column such that each genre comes in each line 
df['listed_in'] = df['listed_in'].str.split(', ')          
df = df.explode('listed_in').reset_index(drop = True)  
```

```python
# Converting into category data type
df['listed_in'] = df['listed_in'].astype('category')
df.head()
```

# Data VIsualization

# Most frequent genre of movies released

```python
sn.catplot(y = 'listed_in', data = df, order = df['listed_in'].value_counts().index, color = 'blue', kind = 'count',   # to get the value counts of each genre in listed_in column
    height=15,      # taller           # index --> to arrange the order based on index (whoever comes first)
    aspect= 0.5   )  # wider
plt.title('genre frequency graph')
plt.show()   #Seaborn automatically chooses tick intervals (0, 1000, 2000, etc.) based on the largest count
```

# frequency of movies released in each year

```python
 # Category plotting 
sn.catplot( y = 'release_year', data = df, order = df['release_year'].value_counts().index, kind = 'count')
plt.title('frequency of movies released in each year')
plt.show()
```
